```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

## Yeo lab test for if peak is significant
## lamda is an estimate for num reads in a peak
## rip is observed num of reads in peak

lam  = 500
rip = 1000

x = np.linspace(0,1500,1500)
cdf = poisson.cdf(x,lam)
sfun = poisson.sf(x,lam)
plt.plot(x,cdf)
plt.plot(x,sfun)
print(poisson.sf(rip,lam))
plt.vlines(rip,0,1,linestyle='--')
```
